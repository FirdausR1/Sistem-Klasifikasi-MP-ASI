```python
import matplotlib.pyplot as plt

# Data
provinsi = ['DKI Jakarta', 'Papua']
persentase = [99.8, 28.0]  # Persentase (sebagai contoh)

# Membuat grafik batang
plt.figure(figsize=(8, 5))
plt.bar(provinsi, persentase, color=['blue', 'orange'])
plt.title('Ketimpangan Pemberian Makanan Tambahan Balita')
plt.xlabel('Provinsi')
plt.ylabel('Persentase (%)')
plt.ylim(0, 100)  # Batas Y
plt.grid(axis='y')

# Menampilkan grafik
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
provinsi = ['DKI Jakarta', 'Papua']
persentase = [99.8, 28.0]  # Persentase (sebagai contoh)

# Membuat pie chart
plt.figure(figsize=(8, 5))
plt.pie(persentase, labels=provinsi, autopct='%1.1f%%', colors=['green', 'orange'])
plt.title('Ketimpangan Pemberian Makanan Tambahan Balita')

# Menampilkan grafik
plt.show()
```
