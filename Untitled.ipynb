```python
from flask import Flask, request, jsonify, render_template
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
import joblib
import os

app = Flask(__name__)


mp_asi_data_path = 'DATASET/Databayi1.csv'
resep_makanan_data_path = 'DATASET/Data_Rekomendasi_Menu.xlsx'

mp_asi_df = pd.read_csv(mp_asi_data_path)
resep_makanan_df = pd.read_excel(resep_makanan_data_path)

resep_makanan_df = resep_makanan_df.dropna()
print(resep_makanan_df.head())

resep_makanan_df['Kalori (kkal)'] = pd.to_numeric(resep_makanan_df['Kalori (kkal)'], errors='coerce')

label_encoder_jenis_kelamin = LabelEncoder()
mp_asi_df['Jenis Kelamin'] = label_encoder_jenis_kelamin.fit_transform(mp_asi_df['Jenis Kelamin'])

label_encoder_tingkat_aktivitas = LabelEncoder()
mp_asi_df['Tingkat Aktivitas'] = label_encoder_tingkat_aktivitas.fit_transform(mp_asi_df['Tingkat Aktivitas'])

if 'Kebutuhan Kalori' not in mp_asi_df.columns:
    mp_asi_df['Kebutuhan Kalori'] = mp_asi_df['Berat Badan (kg)'] * 70  # Contoh perhitungan sederhana

mp_asi_numerik = mp_asi_df.select_dtypes(include=[np.number])
mp_asi_kategori = mp_asi_df.select_dtypes(exclude=[np.number])

# Isi nilai NaN hanya pada kolom numerik
mp_asi_numerik = mp_asi_numerik.fillna(mp_asi_numerik.mean())

# Gabungkan kembali kolom numerik dan kategori
mp_asi_df = pd.concat([mp_asi_numerik, mp_asi_kategori], axis=1)

# Pisahkan kolom numerik dan kategori pada resep_makanan_df
resep_makanan_numerik = resep_makanan_df.select_dtypes(include=[np.number])
resep_makanan_kategori = resep_makanan_df.select_dtypes(exclude=[np.number])

# Isi nilai NaN hanya pada kolom numerik
resep_makanan_numerik = resep_makanan_numerik.fillna(resep_makanan_numerik.mean())

# Gabungkan kembali kolom numerik dan kategori
resep_makanan_df = pd.concat([resep_makanan_numerik, resep_makanan_kategori], axis=1)

X_menu = mp_asi_df[['Tinggi Badan (cm)', 'Berat Badan (kg)', 'Umur Balita (bulan)', 'Jenis Kelamin', 'Tingkat Aktivitas', 'Kebutuhan Kalori']].iloc[:216]
y_menu = resep_makanan_df['Kategori Menu']

# Mengatasi ketidakseimbangan data dengan SMOTE dan n_neighbors lebih rendah
smote = SMOTE(random_state=42, k_neighbors=1)
X_menu_smote, y_menu_smote = smote.fit_resample(X_menu, y_menu)

X_train_menu, X_test_menu, y_menu_train, y_menu_test = train_test_split(X_menu_smote, y_menu_smote, test_size=0.3, random_state=42)

rf_menu_classifier = RandomForestClassifier(random_state=42)

# Grid search untuk hyperparameter tuning
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(rf_menu_classifier, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_menu, y_menu_train)


best_rf_classifier = grid_search.best_estimator_
best_rf_classifier.fit(X_train_menu, y_menu_train)

# Prediksi dan evaluasi model kategori menu
y_menu_pred = best_rf_classifier.predict(X_test_menu)

# Menampilkan akurasi keseluruhan
akurasi_menu = accuracy_score(y_menu_test, y_menu_pred)
print(f"Akurasi Model Kategori Menu: {akurasi_menu * 100:.2f}%")

precision_menu = precision_score(y_menu_test, y_menu_pred, average='weighted')
recall_menu = recall_score(y_menu_test, y_menu_pred, average='weighted')
f1_menu = f1_score(y_menu_test, y_menu_pred, average='weighted')

print(f"Precision Model Kategori Menu (Weighted): {precision_menu * 100:.2f}%")
print(f"Recall Model Kategori Menu (Weighted): {recall_menu * 100:.2f}%")
print(f"F1 Score Model Kategori Menu (Weighted): {f1_menu * 100:.2f}%")
# Menghitung precision, recall, dan F1 secara rinci dengan classification_report
report = classification_report(y_menu_test, y_menu_pred, target_names=np.unique(y_menu).astype(str))
print("Laporan Klasifikasi Model Kategori Menu:")
print(report)

# Menyimpan model
model_path = 'menu_decision_tree_model.pkl'
joblib.dump(best_rf_classifier, model_path)

print(f"Model Random Forest Kategori Menu disimpan di: {model_path}")
```

```python
# 1. Memeriksa apakah ada nilai kosong di kolom Kalori (kkal)
missing_calories = resep_makanan_df['Kalori (kkal)'].isnull().sum()
print(f"Jumlah nilai kosong di kolom Kalori (kkal): {missing_calories}")

# 2. Jika ada nilai kosong, isi dengan nilai median atau rata-rata
if missing_calories > 0:
    # Menggunakan median, karena lebih tahan terhadap outlier
    median_calories = resep_makanan_df['Kalori (kkal)'].median()
    resep_makanan_df = resep_makanan_df.assign(**{'Kalori (kkal)': resep_makanan_df['Kalori (kkal)'].fillna(median_calories)})
    print(f"Nilai kosong di kolom Kalori (kkal) telah diisi dengan nilai median: {median_calories}")

# 3. Pastikan semua nilai di kolom Kalori (kkal) sudah terisi
missing_calories_after = resep_makanan_df['Kalori (kkal)'].isnull().sum()
print(f"Jumlah nilai kosong di kolom Kalori (kkal) setelah imputasi: {missing_calories_after}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

conf_matrix = confusion_matrix(y_menu_test, y_menu_pred)
conf_display = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=best_rf_classifier.classes_)
plt.figure(figsize=(8, 8))
conf_display.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    best_rf_classifier, X_train_menu, y_menu_train, cv=5, n_jobs=-1, scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 10)
)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training accuracy")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Validation accuracy")
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, color="r", alpha=0.1)
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, color="g", alpha=0.1)
plt.title("Learning Curve - Akurasi Model")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend(loc="best")
plt.grid()
plt.show()
```
